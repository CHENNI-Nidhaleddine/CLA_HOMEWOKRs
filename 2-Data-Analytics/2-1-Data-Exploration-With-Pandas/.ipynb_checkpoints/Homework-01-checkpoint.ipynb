### Defining Some functions to calculate the wanted quantities

```python
def getMean(L):
    return sum(L)/len(L)
#-----------------------
def getMedian(L):
    return sorted(L)[int(len(L)/2)]
#-----------------------
def getMode(L):
    X={}
    max=0
    for i in L:
        if i in X:
            X[i]+=1
        else:
            X[i]=1
        if(X[i]>max):
            max=X[i]
            mode=i
    return i
#------------------------
def getVariance(L,mean):
    v=0
    for i in L:
        v+=(i-mean)**2
    return v/len(L)
#------------------------
def getStdDeviation(v):
    return v**0.5
```

```python
mean=getMean(eu_sales)
median=getMedian(eu_sales)
mode=getMode(eu_sales)
variance=getVariance(eu_sales,mean)
stdDeviation=getStdDeviation(variance)
print("mean:",mean,"\nMedian:",median,"\nMode:",mode,"\nVariance:",variance,"\nDeviation:",stdDeviation)
```
