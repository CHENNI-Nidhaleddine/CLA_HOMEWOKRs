```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import sklearn.metrics
import matplotlib.pyplot as plt
```

### Reading data

```python
df=pd.read_csv('Fish.csv')
df.dropna(inplace=True)
```

### Building the lenear Regression Model

```python
model=LinearRegression()
x=df.iloc[:,2:7].values
y=df.iloc[:,2].values
model=model.fit(x,y)
```

### Mesuring the Quality of the model with R**2 Score

```python
y_pred=model.predict(x)
sklearn.metrics.r2_score(y, y_pred,multioutput='uniform_average')
```

### ploting for ensuring

```python
# we got R**2 =1 which mean that the model predict the actual value we will now plot y and y-pred to see if they fits
plt.figure(figsize=(12, 6))
plt.plot(x, y, 'o')           # scatter plot showing actual data
plt.plot(x, y_pred, 'r', linewidth=2)  
```
